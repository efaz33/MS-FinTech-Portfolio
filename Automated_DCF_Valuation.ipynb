```python
import pandas as pd

# --- STAGE 1: FINANCIAL VALUATION INPUTS ---
company_name = "Apple Inc. (Simulation)"
current_revenue = 385000000000  # $385 Billion baseline
fcf_margin = 0.25               # 25% Free Cash Flow margin
expected_growth = 0.08          # 8% annual growth rate for next 5 years
discount_rate = 0.09            # 9% Weighted Average Cost of Capital (WACC)
perpetual_growth = 0.02         # 2% long-term economic growth rate

# --- STAGE 2: 5-YEAR CASH FLOW FORECASTING ---
years = [1, 2, 3, 4, 5]
projected_fcfs = []
present_values = []

for year in years:
    # Estimate future revenue based on compounding growth
    future_revenue = current_revenue * ((1 + expected_growth) ** year)
    future_fcf = future_revenue * fcf_margin
    # Discount future cash flow back to today's money values
    pv_fcf = future_fcf / ((1 + discount_rate) ** year)

    projected_fcfs.append(future_fcf)
    present_values.append(pv_fcf)

# --- STAGE 3: TERMINAL VALUE & INTRINSIC ENTERPRISE VALUE ---
terminal_value = (projected_fcfs[-1] * (1 + perpetual_growth)) / (discount_rate - perpetual_growth)
pv_terminal_value = terminal_value / ((1 + discount_rate) ** 5)
intrinsic_enterprise_value = sum(present_values) + pv_terminal_value

print(f"--- Automated DCF Valuation for {company_name} ---")
print(f"Sum of Present Value of 5-Year Cash Flows: ${sum(present_values):,.2f}")
print(f"Present Value of Terminal Value (Year 6+):  ${pv_terminal_value:,.2f}")
print(f"Calculated Intrinsic Enterprise Value:       ${intrinsic_enterprise_value:,.2f}")
```
