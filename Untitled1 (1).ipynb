```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# 1. Initialize text analytics engine
nltk.download('vader_lexicon', quiet=True)
sia = SentimentIntensityAnalyzer()

# 2. Extract from a live financial news stream (Yahoo Finance)
url = "https://finance.yahoo.com/news/rssindex"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}

try:
    response = requests.get(url, headers=headers, timeout=10)
    # Parse as true XML for RSS feeds
    soup = BeautifulSoup(response.content, features="xml")
    articles = soup.find_all('item')

    market_data = []
    for item in articles[:10]:
        headline = item.title.text if item.title else "No Title"
        pub_date = item.pubDate.text if item.pubDate else "No Date"

        # Calculate quantitative VADER sentiment metrics
        score = sia.polarity_scores(headline)['compound']
        market_data.append({
            "Timestamp": pub_date,
            "Headline": headline,
            "Sentiment_Score": score
        })

    df = pd.DataFrame(market_data)

    print("--- Advanced Algorithmic Sentiment Analysis ---")
    if df.empty:
        print("Pipeline running, but feed payload is currently empty.")
    else:
        # Display the full table cleanly without truncation
        pd.set_option('display.max_colwidth', None)
        print(df[["Timestamp", "Headline", "Sentiment_Score"]].to_string(index=False))

except Exception as e:
    print(f"Data pipeline streaming error: {e}")
```
