```python
import yfinance as yf
import pandas as pd

# 1. Force auto_adjust=False to guarantee the 'Adj Close' column structure exists
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']
raw_data = yf.download(tickers, start="2018-01-01", end="2024-01-01", auto_adjust=False)

# 2. Extract the 'Adj Close' column sub-level directly
adj_close_data = raw_data['Adj Close']

# 3. Calculate daily percentage returns and drop the first empty row
returns = adj_close_data.pct_change().dropna()

print("--- Fixed Successfully! First 5 rows of daily returns: ---")
print(returns.head())
```

```python
import numpy as np
import pandas as pd

# --- Step 2: Calculating Expected Returns, Risk, and the Covariance Matrix ---

# Calculate daily expected returns for each stock
expected_returns_daily = returns.mean()

# Annualize the expected returns
expected_returns_annual = expected_returns_daily * 252

# Calculate the daily covariance matrix
cov_matrix_daily = returns.cov()

# Annualize the covariance matrix
cov_matrix_annual = cov_matrix_daily * 252

print("\nAnnualized Expected Returns:")
print(expected_returns_annual)

print("\nAnnualized Covariance Matrix (first 5x5):")
print(cov_matrix_annual.head())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Step 3: Portfolio Optimization (Modern Portfolio Theory) ---


# Number of portfolios to simulate
num_portfolios = 10000

# Initialize arrays to store results
all_portfolio_returns = []
all_portfolio_std_devs = []
all_portfolio_weights = []

# Set a risk-free rate (e.g., 1% or 0.01)
# For a simplified exercise, we'll assume a risk-free rate of 0 for now to focus purely on portfolio construction.
risk_free_rate = 0.0

# Convert expected_returns_annual and cov_matrix_annual to numpy arrays for easier calculation
expected_returns_array = expected_returns_annual.to_numpy()
cov_matrix_array = cov_matrix_annual.to_numpy()

print(f"Simulating {num_portfolios} random portfolios...")

for i in range(num_portfolios):
    # Generate random weights for each asset
    weights = np.random.random(len(tickers))
    # Ensure weights sum to 1 (100% of the portfolio)
    weights /= np.sum(weights)

    # Calculate portfolio return
    portfolio_return = np.sum(weights * expected_returns_array)

    # Calculate portfolio standard deviation (volatility)
    portfolio_std_dev = np.sqrt(np.dot(weights.T, np.dot(cov_matrix_array, weights)))

    # Store results
    all_portfolio_returns.append(portfolio_return)
    all_portfolio_std_devs.append(portfolio_std_dev)
    all_portfolio_weights.append(weights)

# Convert lists to numpy arrays for easier analysis
portfolio_returns = np.array(all_portfolio_returns)
portfolio_std_devs = np.array(all_portfolio_std_devs)
portfolio_weights = np.array(all_portfolio_weights)

# Calculate Sharpe Ratio for each portfolio
sharpe_ratios = (portfolio_returns - risk_free_rate) / portfolio_std_devs

# Find the portfolio with the maximum Sharpe Ratio
max_sharpe_idx = sharpe_ratios.argmax()
max_sharpe_portfolio_return = portfolio_returns[max_sharpe_idx]
max_sharpe_portfolio_std_dev = portfolio_std_devs[max_sharpe_idx]
max_sharpe_portfolio_weights = portfolio_weights[max_sharpe_idx]

# Find the portfolio with the minimum volatility
min_vol_idx = portfolio_std_devs.argmin()
min_vol_portfolio_return = portfolio_returns[min_vol_idx]
min_vol_portfolio_std_dev = portfolio_std_devs[min_vol_idx]
min_vol_portfolio_weights = portfolio_weights[min_vol_idx]

print("\n--- Optimization Results ---")
print(f"Maximum Sharpe Ratio Portfolio (Risk-Free Rate: {risk_free_rate:.2f}):")
print(f"  Expected Annual Return: {max_sharpe_portfolio_return:.2%}")
print(f"  Annual Volatility (Risk): {max_sharpe_portfolio_std_dev:.2%}")
print(f"  Sharpe Ratio: {sharpe_ratios[max_sharpe_idx]:.2f}")
print("  Optimal Weights:")
for i, ticker in enumerate(tickers):
    print(f"    {ticker}: {max_sharpe_portfolio_weights[i]:.2%}")

print("\nMinimum Volatility Portfolio:")
print(f"  Expected Annual Return: {min_vol_portfolio_return:.2%}")
print(f"  Annual Volatility (Risk): {min_vol_portfolio_std_dev:.2%}")
print("  Optimal Weights:")
for i, ticker in enumerate(tickers):
    print(f"    {ticker}: {min_vol_portfolio_weights[i]:.2%}")

# --- Visualization of the Efficient Frontier ---
plt.figure(figsize=(12, 8))
plt.scatter(portfolio_std_devs, portfolio_returns, c=sharpe_ratios, cmap='viridis', marker='o', alpha=0.5)
plt.colorbar(label='Sharpe Ratio')
plt.title('Portfolio Optimization: Efficient Frontier')
plt.xlabel('Annual Volatility (Standard Deviation)')
plt.ylabel('Expected Annual Return')
plt.grid(True)

# Plot the Maximum Sharpe Ratio portfolio
plt.scatter(max_sharpe_portfolio_std_dev, max_sharpe_portfolio_return, marker='*', color='red', s=500, label='Max Sharpe Ratio Portfolio')

# Plot the Minimum Volatility portfolio
plt.scatter(min_vol_portfolio_std_dev, min_vol_portfolio_return, marker='*', color='blue', s=500, label='Min Volatility Portfolio')

plt.legend()
plt.show()
```
